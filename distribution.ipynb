```python
import seaborn as sns #plotting library
import pandas as pd
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs
%matplotlib inline

df1 = pd.read_csv('steps.csv',sep=';')
df2 = pd.read_csv('survey.csv')
df = pd.merge(df1, df2, on = 'id') #'id' is the common identifier
df.head(10)
```

```python
mean_steps = df.iloc[:,1:328].mean(axis=1) 
mean_steps.index
```

```python
ax = mean_steps.plot() 
ax.set_xlabel('participant id')
ax.set_ylabel('mean steps')
ax.set_title('mean steps over all participants')
```

```python
sns.distplot(df["weight"], hist=False, kde=True)
plt.title("weight distribution of participants")
plt.xlabel("weight")
plt.ylabel("Proportion")
plt.show()
```

```python
sns.histplot(df["weight"], kde=False) #kde=False means we don't get an overlying density plot
plt.title("weight distribution of participants")
plt.xlabel("weight")
plt.ylabel("count")
plt.show()
```
