# Exploring categorical variables

In this Notebook we will explore relations with categorical (qualitative) variables. We will use a data set on our class data about tech. This is a sample of the total data set

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('class.csv')
df.head(50)
```

Which of these are categorical?

 gender (strictly speaking)<br>
 os_phone<br>
 like_social<br>
 like_tech

A simple overview of a single variable can be easily done using Pandas' value_counts.

```python
pd.value_counts(df["gender"])
```

Let's say we are interested to know whether women like tech more. We can make a crosstab of the two variables.

```python
pd.crosstab(df["gender"], df["like_tech"])
```

It's hard to compare the different categories using absolute numbers. Let's use normalized data (in proportions):

```python
pd.crosstab(df["gender"],df["like_tech"], normalize="columns")
```

Hmm, something seems to have gone wrong. We still can't properly compare. What we need is column percentages. For your exercise, look up the documentation for this function and see if you can find out how.

This is a sample, so what conclusion can we draw for the population? In addition to the percentages, we need to take into account sample size. Percentages can easily shift with small numbers, e.g. 3/44 is a big difference from 2/43. So we can't draw any firm conclusions. 

Let's make a bar chart out of this data:

```python
sns.countplot(x="like_tech", hue="gender", data=df)
plt.show()
```

This is a bar chart. In Seaborn this is a type of countplot. In the documentation (linked here) you can find all kinds of options.

The syntax is interesting here. The arguments are the different columns of the Pandas data frame. The same principle is applied in the catplot. This type of plot includes all kinds of ways to investigate the relation between a categorical and a numerical variable.

```python
sns.catplot(x="gender", y="instagram_followers", kind="box", data=df)
plt.title('Followers of both genders in instagram')
plt.show()
```
