Exploring a quantitative variable and its interaction with multiple variables, both quantitative and qualitative.

```python
import seaborn as sns #plotting library
import pandas as pd
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs
%matplotlib inline

#preprocessing steps to see the one combined datafrom
df1 = pd.read_csv('steps.csv',sep=';')
df2 = pd.read_csv('survey.csv')
df = pd.merge(df1, df2, on = 'id') #'id' is the common identifier
df.head(10)
```

## a quantitative variable
Create the variable mean steps per participant.

```python
df['mean_steps'] = df.iloc[:,1:328].mean(axis=1) # hint: as a default .mean() in pandas has skipna=true means exclude NA
# clean data
count_na = df['mean_steps'].isna().sum()
print(count_na)
df = df[df['mean_steps'].notna()]
count_na = df['height'].isna().sum()
print(count_na)
df = df[df['height'].notna()]
count_na = df['weight'].isna().sum()
print(count_na)
df = df[df['weight'].notna()]
df.head(10)
```

## Exploratory data
The relation of meansteps per participant with at least 2 other quantitative variables (how are steps related to weight, height)

```python
#Correlation matrix including all 3 variables
corr = df[['height', 'weight','mean_steps']].corr() 
corr
```

Scatter plots of mean_steps with the 2 other variables (height & weight)

```python
sns.scatterplot(data=df,x='height',y='mean_steps')
plt.xlabel('height') #use matplotlib to refine Seaborn plot
plt.ylabel('meansteps')
plt.title('height and mean steps per participant')
plt.show()
```

```python
sns.scatterplot(data=df,x='weight',y='mean_steps')
plt.xlabel('weight') #use matplotlib to refine Seaborn plot
plt.ylabel('meansteps')
plt.title('weight and mean steps per participant')
plt.show()
```

Let's calculate the correlation, Pearson's r. 

```python
from scipy.stats.stats import pearsonr #We will need the correlation function from the scientific package scipy
print(pearsonr(df['mean_steps'],df['height']))
print(pearsonr(df['mean_steps'],df['weight']))
```

mean_steps and height are vice versa  (r = -0.06)<br>
The correlation is moderate (p < 0.03)

mean_steps and weight have no relationship  (r = -0.01)<br>
The correlation has no certainty (p > 0.6)
